```python
df = np.loadtxt('./handwritten_data_785.csv', delimiter=',')
X = df[:, 1:785]
y = df[:, 0]
```

```python
plt.imshow(X[2000].reshape(28,28), cmap=plt.cm.gray)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
linearModel = LinearRegression()
linearModel.fit(X_train, y_train)
lAccTrain = linearModel.score(X_train, y_train) * 100
lAccTest = linearModel.score(X_test, y_test) * 100
print('Train Accuracy is: {:.3f}%'.format(lAccTrain))
print('Test Accuracy is: {:.3f}%'.format(lAccTest))
```

```python
from sklearn.tree import DecisionTreeClassifier
decisionModel = DecisionTreeClassifier()
decisionModel.fit(X_train, y_train)
dAccTrain = decisionModel.score(X_train, y_train) * 100
dAccTest = decisionModel.score(X_test, y_test) * 100
print('Train Accuracy is: {:.3f}%'.format(dAccTrain))
print('Test Accuracy is: {:.3f}%'.format(dAccTest))
```

```python
from sklearn.ensemble import RandomForestClassifier
forestModel = RandomForestClassifier()
forestModel.fit(X_train, y_train)
fAccTrain = forestModel.score(X_train, y_train) * 100
fAccTest = forestModel.score(X_test, y_test) * 100
print('Train Accuracy is: {:.3f}%'.format(fAccTrain))
print('Test Accuracy is: {:.3f}%'.format(fAccTest))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(forestModel, X_train, y_train, cv=10)
scores.mean
```
